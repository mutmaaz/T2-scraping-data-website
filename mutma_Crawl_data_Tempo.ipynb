```python
# Install library yang dibutuhkan
!pip install feedparser
!pip install pandas
!pip install openpyxl  # untuk export Excel

import feedparser
import pandas as pd
import json

# --- Daftar RSS feed Tempo.co ---
rss_urls = {
    "nasional": "https://rss.tempo.co/nasional",
    "dunia": "https://rss.tempo.co/dunia",
    "bisnis": "https://rss.tempo.co/bisnis",
    "tekno": "https://rss.tempo.co/tekno",
    "bola": "https://rss.tempo.co/bola"
}

all_data = []

for kategori, rss_url in rss_urls.items():
    feed = feedparser.parse(rss_url)
    print(f"Kategori '{kategori}': Menemukan {len(feed.entries)} artikel.")

    for entry in feed.entries:
        all_data.append({
            "Kategori": kategori,
            "Judul": entry.title,
            "Link": entry.link,
            "Tanggal": entry.published
        })

valid_articles = len(all_data)
print(f"Berhasil mengumpulkan {valid_articles} artikel yang valid.")

# --- Simpan ke JSON ---
with open('list_post_tempo_lengkap.json', 'w', encoding='utf-8') as f:
    json.dump(all_data, f, ensure_ascii=False, indent=4)

# --- Simpan ke CSV ---
df = pd.DataFrame(all_data)
df.to_csv('list_post_tempo_lengkap.csv', index=False, encoding='utf-8')

# --- Simpan ke Excel ---
df.to_excel('list_post_tempo_lengkap.xlsx', index=False)

print("Data berhasil diekspor ke list_post_tempo_lengkap.json")
print("Data berhasil diekspor ke list_post_tempo_lengkap.csv")
print("Data berhasil diekspor ke list_post_tempo_lengkap.xlsx")
```
